```python
def tag_tokens(text):
    tokens = word_tokenize(text)      
    tagged = nltk.pos_tag(tokens)
    line_postag = [item[0] + '_' + item[1] for item in tagged]
    line_postag = ' '.join(line_postag)
    return line_postag    
```

```python
def get_tokens(text, stopwords):
    
    text_ngrams = []
    tags = ['NN', 'NNS', 'NNP', 'NNPS', 'JJ', '.']
    input_substring = string.punctuation.replace('.','').replace('-','') + '0123456789'    
    remove_dict = dict.fromkeys(input_substring, ' ')
    table = str.maketrans(remove_dict)
    stemmer = nltk.stem.porter.PorterStemmer()
    tokens = text.split(' ')
    tagged_tokens = []
#     print(tokens)
    tagged_tokens = [x.strip().split('_') for x in tokens if x.strip()!= ""]
#     print(tagged_tokens)
    temp_list1 = [[x[0].lower().translate(table).strip(), x[1]] for x in tagged_tokens]
    temp_list = [stemmer.stem(x[0]) for x in temp_list1 if x[0] != '']
    text_ngrams.append(list(nltk.bigrams(temp_list)))
    text_ngrams.append(list(nltk.trigrams(temp_list)))
#     print(text_ngrams)
#     print(temp_list)
    if text_ngrams == []:
        text_ngrams = [[],[]]
    tokens = [x[0] for x in temp_list1 if x[1] in tags] 
    tokens = [token for token in tokens if token not in stopwords]                
    tokens = [stemmer.stem(token).strip(" -_") for token in tokens]
    tokens = [token for token in tokens if token not in stopwords]
    
    return tokens, text_ngrams
```

```python
def create_adj_map(token_list, count_bigram_dict):
    n = len(token_list)
    token_bigrams = list(nltk.bigrams(token_list))
    
    freq_dict = Counter(token_list)
    freq_dict.pop('.', None)
    
    adj_map = {}
    if n == 1:
        adj_map = {next(iter(token_list)):None}
        return freq_dict, adj_map
    
    for item in token_bigrams:
        if item[0] == '.' or item[1] == '.':
            continue
        if item in count_bigram_dict:
            if item[0] in adj_map:
                if item[1] not in adj_map[item[0]]:
                    adj_map[item[0]][item[1]] = count_bigram_dict[item]
            else:
                adj_map[item[0]] = {}
                adj_map[item[0]][item[1]] = count_bigram_dict[item]
                
            if item[1] in adj_map:
                if item[0] not in adj_map[item[1]]:
                    adj_map[item[1]][item[0]] = count_bigram_dict[item]
            else:
                adj_map[item[1]] = {}
                adj_map[item[1]][item[0]] = count_bigram_dict[item]
        
    return freq_dict, adj_map  
```

```python
def PageRank(s, alpha, adj_map, iter_count):
    iter_count += 1
    n = len(s)
    p = 1/n
    
    for key in s:
        total = 0.0
        if key in adj_map and adj_map[key]:
            for adj_key in adj_map[key]:
                num = (adj_map[adj_key][key] * s[adj_key])
                deno = 0
                for adj_adj_key in adj_map[adj_key]:
                    deno += adj_map[adj_key][adj_adj_key]
                total += (num/deno)
        s[key] = alpha * total + (1-alpha)*p
    if iter_count == 10:
        return
    else:
        PageRank(s, alpha, adj_map, iter_count)
    return s  
```

```python
def add_bigrams(a, bigrams):
    for item in bigrams:
        if(item[0] in a and item[1] in a):
            a[item] = a[item[0]] + a[item[1]]
def add_trigrams(a, trigrams):
    for item in trigrams:
        if(item[0] in a and item[1] in a and item[2] in a):
            a[item] = a[item[0]] + a[item[1]] + a[item[2]]
```

```python
def get_keyphrases_pr(train_df1, alpha):
    pred_list = {}
    for index, doc in train_df1.iterrows():
        text = doc.Title + doc.Abstract 
        adj_map = {}
        freq_dict = {}
        create_adj_map2(freq_dict, adj_map, doc.Title, doc.title_ngrams[0], 'Title')
        create_adj_map2(freq_dict, adj_map, doc.Abstract, doc.Abstarct_ngrams[0], 'Abstract')
#         create_adj_map2(adj_map, doc.Conclusion, doc.Concl_ngrams[0], 'Conclusion')
#         create_adj_map2(adj_map, doc.Rem_Text, doc.Rem_ngrams[0], 'Rem_Text')
#         adj_map = create_adj_map(text, doc.bigrams)

        vocab = set(text)
        if '.' in vocab:
            vocab.remove('.')
        n = len(vocab)
        init_s = {}
        for word in text:
            if word!= '.' and word not in init_s:
                init_s[word] = 1/n
        pr_values = PageRank(init_s, alpha, adj_map, 0)
        add_bigrams(pr_values, doc.bigrams)
        add_trigrams(pr_values, doc.trigrams)
#         print(pr_values)
        sorted_pr_values = sorted(pr_values, key = pr_values.get , reverse = True)
#         print(sorted_pr_values)
        pred_list[doc.filename] = [" ".join(x) if not isinstance(x, str) else x for x in sorted_pr_values if x != '']

    return pred_list
```

```python
def get_doc_MRR(gold_list, pred_list):
    for i in range(len(pred_list)):
        if pred_list[i] in gold_list:
            return 1/(i+1)
    return 0    
```

```python
def get_gloabl_MRR(gold_dict, pred_list, k):
    MRR = np.zeros(k)
    for doc in pred_list:
        for i in range(k):
            MRR[i] += get_doc_MRR(gold_dict[doc], pred_list[doc][:i+1])
    return MRR/len(gold_dict) 
    
```

```python
def display_results(MRR):
    for i in range(len(MRR)):
        print("K = " + str(i+1) + " MRR = " + str(MRR[i])) 
```
